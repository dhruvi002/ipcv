```python
def histStretching(img, rMax, rMin, sMax, sMin):
    stretchedImg = []
    for x in range(img.shape[0]):
        row = []
        for y in range(img.shape[1]):
            pixel = int((((sMax - sMin) / (rMax - rMin)) * (img[x][y] - rMin)) + sMin)
            row.append(pixel)
        stretchedImg.append(row)

    return np.array(stretchedImg)
```

```python
def showHist(img):
  rows, cols = img.shape
  count = {}

  for x in range(rows):
    for y in range(cols):
      if (img[x][y] in count.keys()):
        count[img[x][y]] += 1
      else:
        count.update({img[x][y]:1})
  plt.figure(figsize = (15,7))
  plt.bar(list(count.keys()), list(count.values()))
  plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

def histStretching(img, rMax, rMin, sMax, sMin):
    stretchedImg = []
    for x in range(img.shape[0]):
        row = []
        for y in range(img.shape[1]):
            pixel = int((((sMax - sMin) / (rMax - rMin)) * (img[x][y] - rMin)) + sMin)
            row.append(pixel)
        stretchedImg.append(row)
    return np.array(stretchedImg)

def showHist(img):
    count = {}
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            if (img[x][y] in count.keys()):
                count[img[x][y]] += 1
            else:
                count.update({img[x][y]:1})
    plt.figure(figsize=(15, 7))
    plt.bar(list(count.keys()), list(count.values()))
    plt.show()

# Read an image using OpenCV
image = cv2.imread(path, 0)

# Perform histogram stretching
rMin, rMax = image.min(), image.max()
sMin, sMax = 0, 255
stretched_img = histStretching(image, rMax, rMin, sMax, sMin)

# Display the original and stretched images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(stretched_img, cmap='gray')
plt.title('Stretched Image')

plt.show()

# Display the histogram
showHist(stretched_img)
showHist(img)
```
