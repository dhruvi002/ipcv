```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def conv(img, mask):
  rows, cols = img.shape
  ans = []

  for i in range(rows-2):
    tempRow = []
    for j in range(cols-2):
      pixels = 0
      for k in range(3):
        for l in range(3):
          pixels += img[i+k, j+l]*mask[k,l]
      tempRow.append(pixels)
    ans.append(tempRow)
  return ans
```

```python
img = cv2.imread('paudha.jpeg', 0)
```

```python
plt.imshow(img, cmap = 'gray')
plt.title('Original Image')
plt.axis('off')
```

# Prewitts

```python
PX = np.array([
    [-1,-1,-1],
    [0,0,0],
    [1,1,1]
])
```

```python
PY = np.array([
    [-1,0,1],
    [-1,0,1],
    [-1,0,1]
])
```

```python
P = PX + PY
```

```python
XPImg = conv(img, PX)
YPImg = conv(img, PY)
PImg = conv(img, P)
```

```python
plt.imshow(XPImg, cmap = 'gray')
plt.title('Prewitts X Image')
plt.axis('off')
```

# Sobel

```python
SX = np.array([
    [-1,-2,-1],
    [0,0,0],
    [1,2,1]
])
```

```python
SY = np.array([
    [-1,0,1],
    [-2,0,2],
    [1,2,1]
])
```

```python
S = SX + SY
```

```python
XSImg = conv(img, SX)
YSImg = conv(img, SY)
SImg = conv(img, S)
```

```python
plt.imshow(XSImg, cmap = 'gray')
plt.title('Sobel X Image')
plt.axis('off')
```
