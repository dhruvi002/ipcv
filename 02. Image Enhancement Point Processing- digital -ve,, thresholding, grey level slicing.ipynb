```python
import matplotlib.pyplot as plt
```

```python
def plotimg(*img, title):
  fig = plt.figure(figsize = (15,7))
  cols = len(img)
  for i in range(cols):
    fig.add_subplot(1, cols, i+1)
    plt.imshow(img[i], cmap = 'gray')
    plt.title(title[i])
```

# Image Negatives

```python
def digNeg(img):
  #neg = []
  neg = [[255-y for y in x] for x in img]
  #neg.append(temp)
  plotimg(img, neg, title = ['Original Image','Digital Negative Image'])
#  cv2_imshow(neg)
```

```python
digNeg(img)
```

#Thresholding

```python
def thresh(img, th = 30):
  threshed = [[0 if y < th else 255 for y in x] for x in img]
  plotimg(img, threshed, title = ['Original Image','Thresholded Image'] )
```

```python
thresh(img, 150)
```

#Grey Level Slicing without Background

```python
def noBGgrey(img, lb = 30, ub = 150):
  noBGsliced = [[255 if lb <= y < ub else 0 for y in x] for x in img]
  plotimg(img, noBGsliced, title = ['Original Image','Grey Level Sliced Image without Background'] )
```

```python
noBGgrey(img, 50, 150)
```

#Grey Level Slicing with Background

```python
def BGgrey(img, lb = 30, ub = 150):
  BGsliced = [[255 if lb <= y < ub else 0 for y in x] for x in img]
  plotimg(img, BGsliced, title = ['Original Image','Grey Level Sliced Image with Background'] )
```

```python
BGgrey(img, 125, 175)
```

#Conclusion

###Digital Negative
- inverts the grey levels of an image
-s = 255-r

###Thresholding

- Also known as extreme contrast stretching
- It is achieved by making the first and last slope of contrast stretching = 0 and increasing the centre slope

### Grey Level Slicing
- It is used when we need to highlight a specific range of grey levels
- It is similar to thresholding, except we select the band of grey levels
- We use grey level slicing with background in applications where we need to retain the background in addition to enhancing a particular band of grey levels
