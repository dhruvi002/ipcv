```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('sea_link.jpg', 0)
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
def conv(img, mask):
    rows, cols = img.shape
    ans = []

    for i in range(rows - 2):
        tempRow = []
        for j in range(cols - 2):
            pixel = 0
            for k in range(3):
                for l in range(3):
                    pixel += img[i + k, j + l] * mask[k, l]
            tempRow.append(pixel)
        ans.append(tempRow)

    return np.array(ans)
```

```python
HPfilter = np.array([
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]
    ], dtype=object)
```

```python
HBfilter = np.array([
    [-1, -1, -1],
    [-1, 8.9, -1],
    [-1, -1, -1]
    ], dtype=object)
```

```python
highPassImg = conv(img, HPfilter)
highBoostImg = conv(img, HBfilter)
```

```python
plt.imshow(highPassImg, cmap = 'gray')
```

```python
plt.imshow(highBoostImg, cmap = 'gray')
```
