# Theory

## OpenCV

### Features

* Read and write images.
* Capture and save videos.
* Process images (filter, transform)
* Perform feature detection.
* Detect specific objects such as faces, eyes, cars, in the videos or images.
* Analyze the video, i.e., estimate the motion in it, subtract the background, and track objects in it.

### Limitations

* Illumination - If there is a slight change in lighting conditions, it will make major impact on its results. It is the lighting to vary, and then the result may be different for the same object cause of low or high illumination.

* Background - The result might not the same outdoor as compared to what is produces indoors because the factor - affecting its performance-change as soon as the locations change.

* Pose - The facial recognition system is highly sensitive to pose variations. The movement of head or different camera positions can cause changes of facial texture and it will generate the wrong result.

* Occlusion - Occlusion means the face as beard, mustache, accessories (goggles, caps, mask, etc.) also interfere with the estimate of a face recognition system.

* Expressions - Change in facial expressions may produce a different result for the same individual.

## Pillow

### Features

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

### Reading Image

```python
image = cv2.imread('gfg.png', cv2.IMREAD_COLOR)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Displaying Image

```python
cv2.imshow('Sample Image', image)
```

```python
plt.title('Sample Image')
plt.imshow(image)
plt.show()
```

### Convert into Gray Scale

```python
greyimg = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('Grey Scaled Image', greyimg)
```

```python
plt.title('Grey Scaled Image')
plt.imshow(greyimg)
plt.show()
```

### Crop Image

```python
print(image.shape)
```

```python
cropped_image = image[80:280, 150:330]
```

```python
cv2.imshow("Cropped Image", cropped_image)
```

```python
plt.title('Cropped Image')
plt.imshow(cropped_image)
plt.show()
```

### Arithmetic Operations on Image

```python
img1 = cv2.imread('input1.jpg') 
img2 = cv2.imread('input2.jpg')
```

#### Weighted Sum

```python
weightedSum = cv2.addWeighted(img1, 0.5, img2, 0.4, 0)
```

```python
cv2.imshow('Weighted Image', weightedSum)
```

```python
plt.title('Weighted Image')
plt.imshow(weightedSum)
plt.show()
```

#### Subtraction

```python
sub = cv2.subtract(img1, img2)
```

```python
cv2.imshow('Subtracted Image', sub)
```

```python
plt.title('Subtracted Image')
plt.imshow(sub)
plt.show()
```

### Logical Operations

#### Bitwise AND

```python
bitwise_and = cv2.bitwise_and(img2, img1)
```

```python
cv2.imshow("Bitwise AND Image", bitwise_and)
```

```python
plt.title('Bitwise AND Image')
plt.imshow(bitwise_and)
plt.show()
```

#### Bitwise OR

```python
bitwise_or = cv2.bitwise_or(img2, img1)
```

```python
cv2.imshow("Bitwise OR Image", bitwise_or)
```

```python
plt.title('Bitwise OR Image')
plt.imshow(bitwise_or)
plt.show()
```

#### Bitwise XOR

```python
bitwise_xor = cv2.bitwise_xor(img2, img1)
```

```python
cv2.imshow("Bitwise XOR Image", bitwise_xor)
```

```python
plt.title('Bitwise XOR Image')
plt.imshow(bitwise_xor)
plt.show()
```
