#Implementation

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
path ='/content/drive/MyDrive/DJ/Sem 6/IPCV/ghoda.jpg'
```

```python
img = cv.imread(path, 0)
cv2_imshow(img)
```

```python
def equalizeHistogram(img):
    rows, cols = img.shape
    histogram = np.zeros(256) 
    cdf = np.zeros(256)

    for i in range(rows):
        for j in range(cols):
            histogram[img[i, j]] +=1

    pdf = histogram / (rows * cols)
    cdf = np.cumsum(pdf)
    final_CDF = np.round(cdf * 255)
    ans = final_CDF[img]
                
    return ans
```

```python
nayaghoda = equalizeHistogram(img)
```

```python
cv2_imshow(nayaghoda)
```

```python
def showHist(img):
  count = {}
  for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      if (img[x][y] in count.keys()):
        count[img[x][y]] += 1
      else:
        count.update({img[x][y]:1})
  plt.figure(figsize = (15,7))
  plt.bar(list(count.keys()), list(count.values()))
  plt.show()
```

```python
print('\t\t\tOriginal Histogram')
showHist(img)
```

```python
print('\t\t\tEqualised Histogram')
showHist(nayaghoda)
```
